```python
# Conexión a Supabase para la data cruda
# Capa Bronze

from pyspark.sql import SparkSession

# Configuración de la DB
db_config = {
    "host": "aws-1...",
    "port": "---",
    "database": "----",
    "user": "---",
    "password": "---" 
}

jdbc_url = f"jdbc:postgresql://{db_config['host']}:{db_config['port']}/{db_config['database']}"

source_tables = ["ordenes", "order_items", "logistica_investigacion", "productos"]

def cargar_bronze(table_name):
    # Función para mover datos de Postgres al lh
    print(f"Cargando tabla: {table_name}")
    try:
        df = spark.read \
            .format("jdbc") \
            .option("url", jdbc_url) \
            .option("dbtable", table_name) \
            .option("user", db_config['user']) \
            .option("password", db_config['password']) \
            .option("driver", "org.postgresql.Driver") \
            .load()
        
        nombre_destino = f"bronze_{table_name}"
        df.write.format("delta").mode("overwrite").saveAsTable(nombre_destino)
        
        print(f"Cargada: {nombre_destino} con {df.count()} filas")
        
    except Exception as e:
        # Si falla la autenticación, aquí te lo dirá
        print(f"Fallo en {table_name}: {str(e)}")

# Ejecución
for table in source_tables:
    cargar_bronze(table)

print("Bronze terminado")
```
